## Kazer dataset:

```python
pacmap_70 = pd.read_csv('../fig_data/kazer_pacmap_70PCs.csv')
pacmap_70 = np.array(pacmap_70)

tsne_70 = np.load('../fig_data/kazer_tsne_70PCs.npy')[0,:,:]

umap_70 = pd.read_csv('../fig_data/kazer_umap_70PCs.csv')
umap_70 = np.array(umap_70)

trimap_70 = pd.read_csv('../fig_data/kazer_trimap_70PCs.csv')
trimap_70 = np.array(trimap_70)
```

```python
label_csv = pd.read_csv('../fig_data/kazer_label.csv')
label_np = np.array(label_csv['x'])
```

```python
color_dict = {'NK cell': 6, 'DCs': 7, 'T cell': 1, 'B cell': 4, 'CTLs': 5, 'monocyte': 2, 'plasmablast': 3}
```

```python
color = np.zeros(label_np.shape)
for i in range(color.shape[0]):
    color[i] = color_dict[label_np[i]]
```

```python
plt.figure(figsize=(10,10))
scatter = plt.scatter(pacmap_70[:, 0], pacmap_70[:, 1], c=color, s=0.5, cmap=plt.cm.get_cmap("RdYlGn_r"), label=label_np)

a = 2.9
b = 2.1
x = np.arange(-14 + a, 14 + a, 0.1)
y = np.arange(-14 + b, 14 + b, 0.1)
x, y = np.meshgrid(x,y)
plt.contour(x, y, 0.4*(x-a)**2 + 1.5*(y-b)**2, [9], colors='black', linewidths=5.5)

plt.arrow(0.6, 6.6, 0, -1.4, width=0.2, head_width=1, color = 'blue')
plt.arrow(3, -2.4, 0, 1.4, width=0.2, head_width=1, color = 'red')

plt.text(-1.5, 7., 'pDCs', color='blue', size=30.0)
plt.text(1., -3.6, 'mDCs', color='red', size=30.0)

plt.title('PaCMAP', fontsize=50)

plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.tight_layout()
plt.savefig('subfig/kazer_pacmap_70_mark')
plt.show()
```

## Stuart dataset:

```python
pacmap_70 = pd.read_csv('../fig_data/stuart_pacmap_70PCs.csv')
pacmap_70 = np.array(pacmap_70)

tsne_70 = np.load('../fig_data/stuart_tsne_70PCs.npy')[0,:,:]

umap_70 = pd.read_csv('../fig_data/stuart_umap_70PCs.csv')
umap_70 = np.array(umap_70)

trimap_70 = pd.read_csv('../fig_data/stuart_trimap_70PCs.csv')
trimap_70 = np.array(trimap_70)
```

```python
label_csv = pd.read_csv('../fig_data/stuart_label.csv')
label_np = np.array(label_csv['x'])

df = pd.DataFrame(label_np, columns=['status'])
status_dict = df['status'].unique().tolist()

df['transformed'] = df['status'].apply(lambda x : status_dict.index(x))
label = np.array(df['transformed'])

simplied_label = np.zeros((label.shape), dtype=np.int32)
for i in range(label.shape[0]):
    if label[i] == 0:
        simplied_label[i] = 0
    elif label[i] == 20:
        simplied_label[i] = 1
    else:
        simplied_label[i] = 2
```

```python
from matplotlib.lines import Line2D

cmap=plt.cm.get_cmap("gist_ncar")
lw=14.0

custom_lines = [Line2D([0], [0], color='gold', lw=lw, label='Prog_RBC'),
                Line2D([0], [0], color='black', lw=lw, label='Prog_MK'),
                Line2D([0], [0], color='gainsboro', lw=lw, label='others')]

plt.figure(figsize=(3,8))
plt.legend(custom_lines, ['Prog_RBC', 'Prog_MK', 'others'], fontsize=25, markerscale=5.0,
                    loc="center", title="Cell Type", title_fontsize=36, labelspacing=1., borderpad=0.4)
plt.axis('off')
#plt.tight_layout()
plt.savefig('subfig/stuart_legend')
plt.show()
```

```python
c_dict = ['gold', 'black', 'gainsboro']
color = []
for i in range(label.shape[0]):
    if simplied_label[i] == 0:
        color.append(c_dict[0])
    elif simplied_label[i] == 1:
        color.append(c_dict[1])
    else:
        color.append(c_dict[2])
```
